# Data Exploration — Cardio AI Platform

This notebook performs exploratory data analysis (EDA) on three cardiovascular datasets:
1. **Heart Disease Dataset** — Clinical indicators for heart disease detection
2. **Cardiac Failure Dataset** — Patient records for cardiac failure prediction
3. **Framingham Dataset** — Longitudinal data for 10-year CHD risk estimation

We also briefly explore the ECG signals dataset used for anomaly detection.

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

print("Libraries loaded successfully")
```

## 1. Load Raw Datasets

```python
# Load datasets
heart_df = pd.read_csv("../data/raw/heart_dataset.csv")
cardiac_df = pd.read_csv("../data/raw/cardiac_failure.csv")
framingham_df = pd.read_csv("../data/raw/framingham.csv")

print(f"Heart Dataset:       {heart_df.shape[0]} rows x {heart_df.shape[1]} cols")
print(f"Cardiac Failure:     {cardiac_df.shape[0]} rows x {cardiac_df.shape[1]} cols")
print(f"Framingham Dataset:  {framingham_df.shape[0]} rows x {framingham_df.shape[1]} cols")
```

## 2. Heart Disease Dataset — Overview

```python
print("=== Heart Dataset Info ===")
print(heart_df.info())
print("\n=== First 5 Rows ===")
heart_df.head()
```

```python
heart_df.describe()
```

### 2.1 Missing Values & Target Distribution

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Missing values
sns.heatmap(heart_df.isnull(), cbar=True, yticklabels=False, ax=axes[0], cmap="viridis")
axes[0].set_title("Missing Values — Heart Dataset")

# Target distribution
heart_df["HeartDisease"].value_counts().plot.pie(
    autopct="%1.1f%%", colors=["#2ecc71", "#e74c3c"],
    labels=["No Disease", "Disease"], ax=axes[1]
)
axes[1].set_title("Heart Disease Distribution")
axes[1].set_ylabel("")

plt.tight_layout()
plt.show()
```

### 2.2 Correlation Heatmap — Heart Dataset

```python
plt.figure(figsize=(14, 10))
numeric_heart = heart_df.select_dtypes(include=[np.number])
sns.heatmap(numeric_heart.corr(), annot=True, fmt=".2f", cmap="RdBu_r", center=0, square=True)
plt.title("Correlation Matrix — Heart Disease Dataset")
plt.tight_layout()
plt.show()
```

```python
print("=== Cardiac Failure Info ===")
print(cardiac_df.info())
print("\n=== First 5 Rows ===")
cardiac_df.head()
```

```python
print("=== Framingham Info ===")
print(framingham_df.info())
print("\n=== First 5 Rows ===")
framingham_df.head()
```

## 5. Key Observations

### Heart Dataset
- Target: `HeartDisease` — binary classification
- Key features: Age, RestingBP, Cholesterol, MaxHR, Oldpeak, chest pain type
- Contains both numeric and one-hot encoded categorical variables

### Cardiac Failure Dataset
- Target: `cardio` — binary classification
- Largest dataset (~70K rows)
- Features include vitals (BP, cholesterol, glucose) and lifestyle (smoking, alcohol, activity)
- Age appears to be normalized (0-1 range)

### Framingham Dataset
- Target: `TenYearCHD` — 10-year coronary heart disease risk
- Based on the well-known Framingham Heart Study
- Includes demographic, behavioral, and medical risk factors
- Has some missing values that need imputation

### ECG Signals
- High-dimensional time-series data (4000+ signal points per sample)
- Used for anomaly detection with Isolation Forest
- Not used for supervised classification
