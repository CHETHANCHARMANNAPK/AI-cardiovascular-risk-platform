# Model Experiments — Cardio AI Platform

This notebook trains and evaluates ML models for all three prediction tasks:
1. Heart Disease — RandomForestClassifier
2. Cardiac Failure — GradientBoostingClassifier with Pipeline
3. Framingham CHD Risk — RandomForestClassifier

We compare accuracy, generate classification reports, and save the best models.

```python
import pandas as pd
import numpy as np
import joblib
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc
import warnings
warnings.filterwarnings("ignore")

os.makedirs("../models", exist_ok=True)
print("Libraries loaded")
```

## 1. Heart Disease Model

```python
# Load and split heart data
heart_df = pd.read_csv("../data/processed/heart_clean.csv")
y_heart = heart_df["HeartDisease"]
X_heart = heart_df.drop("HeartDisease", axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_heart, y_heart, test_size=0.2, random_state=42)

# Train RandomForest
heart_model = RandomForestClassifier(n_estimators=200, random_state=42)
heart_model.fit(X_train, y_train)
heart_pred = heart_model.predict(X_test)

print("Heart Disease Model Accuracy:", accuracy_score(y_test, heart_pred))
print("\nClassification Report:")
print(classification_report(y_test, heart_pred))

# Confusion Matrix
plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, heart_pred), annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix — Heart Disease")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Save model
joblib.dump(heart_model, "../models/heart_model.pkl")
print("Heart model saved")
```

## 2. Cardiac Failure Model

```python
# Load and split cardiac data
cardiac_df = pd.read_csv("../data/processed/cardiac_clean.csv")
y_cardiac = cardiac_df["cardio"]
X_cardiac = cardiac_df.drop("cardio", axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_cardiac, y_cardiac, test_size=0.2, random_state=42)

# Train GradientBoosting with Pipeline
cardiac_model = Pipeline([
    ("scaler", StandardScaler()),
    ("classifier", GradientBoostingClassifier(
        n_estimators=400, learning_rate=0.05, max_depth=3, random_state=42
    ))
])
cardiac_model.fit(X_train, y_train)
cardiac_pred = cardiac_model.predict(X_test)

print("Cardiac Failure Model Accuracy:", accuracy_score(y_test, cardiac_pred))
print("\nClassification Report:")
print(classification_report(y_test, cardiac_pred))

plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, cardiac_pred), annot=True, fmt="d", cmap="Reds")
plt.title("Confusion Matrix — Cardiac Failure")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

joblib.dump(cardiac_model, "../models/cardiac_model.pkl")
print("Cardiac model saved")
```

## 3. Framingham CHD Risk Model

```python
# Load and split framingham data
fram_df = pd.read_csv("../data/processed/framingham_clean.csv")
y_fram = fram_df["TenYearCHD"]
X_fram = fram_df.drop("TenYearCHD", axis=1)

X_train, X_test, y_train, y_test = train_test_split(X_fram, y_fram, test_size=0.2, random_state=42)

# Train RandomForest
fram_model = RandomForestClassifier(n_estimators=200, random_state=42)
fram_model.fit(X_train, y_train)
fram_pred = fram_model.predict(X_test)

print("Framingham Model Accuracy:", accuracy_score(y_test, fram_pred))
print("\nClassification Report:")
print(classification_report(y_test, fram_pred))

plt.figure(figsize=(6, 5))
sns.heatmap(confusion_matrix(y_test, fram_pred), annot=True, fmt="d", cmap="Greens")
plt.title("Confusion Matrix — Framingham Risk")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

joblib.dump(fram_model, "../models/framingham_model.pkl")
print("Framingham model saved")
```

## 4. Model Comparison Summary

```python
# Summary table
summary = pd.DataFrame({
    "Model": ["Heart (RandomForest)", "Cardiac (GradientBoosting)", "Framingham (RandomForest)"],
    "Algorithm": ["RandomForestClassifier", "GradientBoostingClassifier", "RandomForestClassifier"],
    "Estimators": [200, 400, 200],
    "Saved As": ["heart_model.pkl", "cardiac_model.pkl", "framingham_model.pkl"]
})

print("=" * 60)
print("MODEL SUMMARY")
print("=" * 60)
print(summary.to_string(index=False))
print("\nAll models saved to ../models/")
```
